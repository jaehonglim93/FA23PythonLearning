## Lambda

A lambda function is a **small anonymous** function.

A lambda function can take any number of arguments, but can only have one expression.


Format -> lambda (arguments) : (expression)

````python
# We define anonymous function that adds 1 to a number
# Then designate add_one to execute the role of that function
add_one = lambda a : a + 1

add_one(5)
````

````python
# We can take multiple inputs for lambda function

add_three = lambda a, b, c: a + b + c

add_three(700, 300 , 9)
````

We can know why we use lambda by using it inside a normal function.

````python
def multiplier(n):
    return lambda a: a * n
````

````python
times_two = multiplier(2)

times_two(17)
````

````python
times_baek = multiplier(100)

times_baek(456)
````

As we can see, lambda function is useful when we need to use a function momentarily.

## Map

We use map() function to find out multiple result cases for different arguments for a function.

Format -> map(function, list)

````python
one_to_five = [1,2,3,4,5]

def double(x):
  return x + x

result = map(double, one_to_five)

# We need to convert 'result' to list in order to print!
list(result)
````

We can use lambda instead of function part in the map.

````python
list(map(lambda a: a + a, range(1,6)))
````

We can use **filter** instead of map when the return value of a function is boolean.

````python
list(filter(lambda a: a < 5, range(1, 11)))
````

## Class

Python is an object oriented programming language (not fully due to lack of encapsulation).

Almost everything in Python is an object, with its properties and methods.

A Class is like an object constructor, or a "blueprint" for creating objects.

````python
# has_tail is a property/attribute of a class called Pet.
class Pet:
  has_tail = True
````

**Instance** is an example of class.

To explain, think of class as a cookie cutter and instance as a cookie created with that cutter.

````python
# Poppi is an instance of Pet class
Poppi = Pet()

# Baekgu is also an instance of Pet class
Baekgu = Pet()

print(Poppi.has_tail)
print(Baekgu.has_tail)
````

This is the most basic form of class.

We should use
```
_init_
```
to initialize(assign values) to the data members of the class when an object of the class is created.








Format
```
def __init__(self, something, something2, ...):
  self.property1 = something self.property2 = something2 ...
```



````python
class Pet:
  # self represents the instance of the class
  def __init__(self, name, pet_type, has_tail):
      self.name = name
      self.type = pet_type
      self.has_tail = has_tail
````

````python
p1 = Pet("Poppi", "Puppy", True)
p2 = Pet("Baekgu", "Puppy", True)
````

````python
print(p1.name)
print(p2.type)
````

**Method** is a function that can only be used inside of the function.

We can just define an ordinary function inside the class!

````python
class Pet:
  def __init__(self, name, pet_type, has_tail):
    self.name = name
    self.type = pet_type
    self.has_tail = has_tail

  # We have to put 'self' as an input to method
  def talk(self):
    if self.type == "Puppy":
        print("Bark!")
    elif self.type == "Kitty":
        print("Meow!")
````

````python
p1 = Pet("Poppi", "Puppy", True)
p1.talk()
````

````python
p2 = Pet("Nabi", "Kitty", True)
p2.talk()
````

If we want to change the value of the property...

````python
# Poppi is a cat..?
p1.type = "Kitty"

p1.talk()
````

To delete property or instance, we can use **del** command.

````python
# Goodbye Nabi
del p2.name

print(p2.name)
````

````python
# Goodbye nameless cat
del p2

print(p2)
````
