## Inheritance

During last week's session, we learnt about...

```python
class Person():
  def __init__(self, fname, lname):
      self.first = fname
      self.last = lname

  def choolseok(self):
      print("My name is " + self.first + " " + self.last)

p1 = Person("Jaehong", "Lim")
p1.choolseok()
```

Inheritance allows us to define a class that inherits all the methods and properties from another class.



**Parent** class is the class being inherited from, also called base class.

**Child** class is the class that inherits from another class, also called derived class.

```python
# Student Class inherits Person Class
class Student(Person):
  pass
```

```python
p2 = Student("Hangoo", "Kang")
p2.choolseok()
```

We can create another init function in child class.

At this time, we do not use parent class's init function.

```python
class Student(Person):

  def __init__(self, fname, lname):

    self.first = fname
    self.last = lname
```

```python
p3 = Student("Hoyoung", "Min")
p3.choolseok()
```

If we want to use init function of parent class,

```python
class Student(Person):

  def __init__(self, fname, lname):
    Person.__init__(self, fname, lname)
```

We can also use **super()** instead of Person to access the parent class.

```python
class Student(Person):

  def __init__(self, fname, lname):
    super().__init__(self, fname, lname)
```

```python
class Student(Person):

  def __init__(self, fname, lname, year):
    Person.__init__(self, fname, lname)
    self.year = year
```

We can create methods in child class, too!

Child class can use parent class's method.

Parent class cannot use child class's method.

```python
class Person():
  def __init__(self, fname, lname):
    self.first = fname
    self.last = lname

  def choolseuk(self):
    print("My name is " + self.first + " " + self.last)

class Student(Person):
  def __init__(self, fname, lname, year):
    Person.__init__(self, fname, lname)
    self.year = year

  def what_year(self):
    print("I'm in my " + self.year + " year")
```

```python
s1 = Student("Jaehong", "Lim", "Junior")

s1.choolseuk()
s1.what_year()
```

```python
s2 = Person("Hangoo", "Kang")
s2.choolseuk()
s2.what_year()
```

```python
def isPalindrome(n):

  ### YOUR CODE HERE ###
  return n == n[::-1]

  ### END OF CODE ###
```
