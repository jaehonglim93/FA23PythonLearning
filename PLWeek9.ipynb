# Data Visualization

## Introduction

**Pandas** is a Python library for data manipulation and analysis.

To use pandas, we need to write this code snippet!

```python
# We use alias 'pd' when using pandas
import pandas as pd
```

**Matplotlib** is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

To use matplotlib, we need to write this code snippet!

```python
# We use alias 'plt' when using matplotlib
from matplotlib import pyplot as plt
```

There are various ways in which data is provided, but when exchanged as a pure file, it is mainly saved in csv (comma separated variables) format.

You can read the file using pandas' **read_csv** function.

```python
# we are reading aapl.csv file
apple = pd.read_csv("https://raw.githubusercontent.com/jaehonglim93/dataset/main/AAPL.csv")
apple
```

To access information in a csv file, you can use the **iloc** function, similar to the commonly used list data access!

## Using pandas...

```python
# data in 5th row, 1st column
apple.iloc[4, 0]
```

```python
# data in 2nd to 4th row, 4th column
apple.iloc[1:4, 3]
```

```python
# data in 2nd to 4th row, 1st to 4th column
apple.iloc[1:4, 0:4]
```

```python
# data in last to last + 4th row, last column
apple.iloc[-4:, -1]
```

In addition to connecting using coordinates, you can also connect using **Column type name**!

```python
# getting all data under 'Date' column
apple['Date']
```

Lastly, using **loc**, you can use column name in regular iloc!

```python
# data in first row, under column 'High'
apple.loc[1, 'High']
```

We can use **mean()** function to calculate the mean of data in the given list.

```python
# Mean of data from 2nd to 5th row, under third column.
apple.iloc[1:4, 3].mean()
```

To filter output data, we can...

```python
# Inside the bracket that is used for indexing, we can put condition.
apple[apple['Close'] > 170]
```

```python
apple.info()
```

```python
apple.head()
```

Can you tell the difference between the two code snippets below?

```python
apple[(apple['Close'] > 110) & (apple['High'] < 115)]['Date']
```

```python
apple[(apple['Close'] > 110) | (apple['High'] < 115)]['Date']
```

## Data visualization

By using the matplotlib library, we can create graphs!

```python
# Plotting values under column 'Close' as lines
plt.plot(apple['Close'])
```

If you set the graph to match a given date (with the date on the x-axis), a visual error will occur.

```python
plt.plot(apple['Date'], apple['Close'])
```

```python
print(type(apple['Date'].iloc[0]))
# Transform the date (str) in csv into a format that pandas can interpret (timestamp).
apple['Date'] = pd.to_datetime(apple['Date'])
print(type(apple['Date'].iloc[0]))
```

Let's look at the functions of matplotlib.pyplot while looking at the detailed process of creating a graph.

```python
# Setting the size of the graph
plt.figure(figsize = (20, 10))

# Creating a graph by setting the x and y axes
plt.plot(apple['Date'], apple['Close'])

# Creating a title to be placed at the top of the graph
plt.title('Apple Stock Closing Price')

# Specify a name for the x-axis of the graph
plt.xlabel("Date")

# Specify a name for the y-axis of the graph
plt.ylabel("Stock Price($)")
```

Now let’s add another graph and compare!

```python
google = pd.read_csv('https://raw.githubusercontent.com/jaehonglim93/dataset/main/GOOG.csv')

# Seems familiar?
google['Date'] = pd.to_datetime(google['Date'])
```

```python
plt.figure(figsize = (10, 5))
plt.plot(google['Date'], google['Close'])
plt.title('Google Stock Closing Price')
plt.xlabel("Date")
plt.ylabel("Stock Price($)")
```

```python
plt.figure(figsize = (10, 5))
plt.plot(apple['Date'], apple['Close'], label = "Apple")

# adding to existing plot
plt.plot(google['Date'], google['Close'], label = "Google")
plt.title('Apple vs Google Stock Closing Price')
plt.xlabel("Date")
plt.ylabel("Stock Price($)")
plt.legend()
```

Lastly, if you want to add a new term to the existing file through calculation, you can use **to_list()** function

```python
a_close = apple['Close'].to_list()
g_close = google['Close'].to_list()

a_start, g_start = a_close[0], g_close[0]

a_close = [(i - a_start)/a_start * 100 for i in a_close]
g_close = [(i - g_start)/g_start * 100 for i in g_close]

# renaming
apple['Close Scaled'] = a_close
google['Close Scaled'] = g_close

apple
```
