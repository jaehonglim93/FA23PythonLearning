## Tuple

**Tuple**, just like list, is a collection of objects. But, unlike list, it is **immutable**. It means that we cannot change, add, or delete the value inside after once defined.

```python
# This is an example of using list

iphone_list = ['i', 'p', 'h', 'o', 'n', 'e']
iphone_list.append('15')
iphone_list[0] = 'I'
print(iphone_list)
```

```python
# This is the same example as above but we used tuple instead of list
iphone_tuple = ('i', 'p', 'h', 'o', 'n', 'e')
iphone_tuple.append('15')
```

```python
# So...would this work?
iphone_tuple[0] = 'I'
```

Other than this difference, tuple shares major features with list

```python
# We can use the len() function
print("The length of iphone_tuple is", len(iphone_tuple))

# We can use index to search items
print("The third element of iphone_tuple is", iphone_tuple[2])

# That includes slicing
print("All odd-indexed elements of iphone_tuple are", iphone_tuple[1::2])

# We can mix different data types
mixed_tuple = ('iphone15', 799, True)

# Can use loop
name = ""
for i in iphone_tuple:
    name += i
print(name)

# Can use if statement
if 'p' in iphone_tuple:
    print("There is p")

# Wait...can we have one element in a tuple?
bad_one_tuple = ('iphone15')
print(type(bad_one_tuple))

# We need to put ',' at the end to make a tuple with one element
good_one_tuple = ('iphone15',)
print(type(good_one_tuple))
```

If you want to change the value inside tuple,

1. Change it to list
2. Change the value of the list
3. Change back to tuple

```python
iphone_tuple_2 = ('i', 'p', 'h', 0, 'n', 'e')

# Change it to list
iphone_tuple_2 = list(iphone_tuple_2)

# Change the value of the list
iphone_tuple_2[3] = 'o'

# Change back to tuple
iphone_tuple_2 = tuple(iphone_tuple_2)

print(iphone_tuple_2)
```

But why do we use tuple?

1. Faster: It takes less time to search inside tuple.
2. Safer: The data inside tuple cannot be corrobated.
3. Dictionary Usage: This will be explained later on.

## Set

**Set** is similar with tuple, but it is **not ordered**.

```python
# Empty Set
companies = {}

companies = {'LG', 'Samsung', 'Foxconn'}

# Run - Runtime - Restart Runtime - Run...Can you see the difference?
print(companies)
```

So, we cannot search data inside a set

```python
print(companies[2])
```

We cannot modify existing values inside the set, but we can add or delete values.

```python
# We use 'add()' instead of 'append()'
companies.add('3M')
print(companies)

companies.remove('LG')
print(companies)
```

Set cannot have a **duplicate** data inside

```python
# How many times 'LG' appears in output?
companies = {'LG', 'Samsung', '3M', 'Foxconn', 'LG'}
print(companies)
```

```python
# Did you see these before???

print("The length of companies is", len(companies))

mixed_set = {'Samsung', 1, True}

for i in companies:
    print(i)

if 'SK' not in companies:
    print("Why not?")
```

We can also get **intersection** and **union** when using set!

```python
companies_1 = {'AMD', 'NVIDIA', 'Intel'}
companies_2 = {'NVIDIA', 'Meta', 'Google'}

print(companies_1.union(companies_2))
print(companies_1.intersection(companies_2))
```

So why do we use set?

1. We can **efficiently** remove duplicates

## Dictionary

Dictionaries are used to store data values in **key:value** pairs, just like actual dictionary.

```python
# Remember the structure of {key:value}!!

# Empty Dictionary (Same as set)
basic_information = {}

# 'Name' -> Key, 'Jaehong Lim' -> value
basic_information["Name"] = "Jaehong Lim"

# Can have multiple keys and multiple values.
basic_information = {
    "Name" : 'Jaehong Lim',
    "Age" : 22,
    "Major" : 'Computer Science',
    "Married" : False
 }

print(basic_information)
```

We have to use key in order to change or add data in dictionary.

```python
# Accessing data stored in 'value' with 'key'
print(basic_information["Major"])

# Changing data stored in 'value' with 'key'
basic_information["Age"] = 21

# Adding data with new 'key' and corresponding new 'value'
basic_information["Address"] = "Seoul"

# We can change data by update() method
basic_information.update({"Address": "Champaign"})

print(basic_information)
```

Dictionary is **ordered**.

Dictionary **cannot** have duplicate keys.

Dictionary **can** have duplicate values.

```python
# How many times 'Year' appear? How many times 'UNFORGIVEN' apperar?
music_info = {
    'Title': 'ANTIFRAGILE',
    'Artist': 'LE SSERAFIM',
    'Album': 'UNFORGIVEN',
    'Year': 2023,
    'Year': "UNFORGIVEN"}

print(music_info)
```

We can use **keys()** and **values()** function to print out separate list of keys and values

```python
print(list(basic_information.keys()))
print(list(basic_information.values()))
```

Wait a minute... How can we use if/ for/ while with dictionary?

```python
# Can you tell the difference between these two?
for i in basic_information:
  if basic_information[i] == 22:
    print("Liar!")

for i in basic_information.keys():
  if basic_information[i] == 22:
    print("Seriously?")
```

```python
# To search for keys...
if "Name" in basic_information:
    print("We have name!")
```

We cannot have list as a key in dictionary because its value may change. However, we can have **tuple**! Remember why?

We can have list and tuple as a value.

```python
# List as value
meal_pref = {'Chinese' : ['Jaehong', 'Hangoo'], 'Japanese': ['Max, June'], 'Korean': ['Seunghwan', 'Joseph']}
meal_pref['Chinese'][0] = 'Danny'

# Tuple as value
meal_pref = {'Chinese' : ('Jaehong', 'Hangoo'), 'Japanese': ('Max, June'), 'Korean': ('Seunghwan', 'Joseph')}

print(meal_pref['Chinese'][0])
```

We can have dictionary inside of a dictionary!

```python
# Inside dictionary called 'students', we have three more separate dictionaries
students = {
    'Jaehong': {
        'Role' : 'Team Leader',
        'Major': 'Computer Science',
        'Team' : 'Python Learning'
    },

    'Hangoo': {
        'Role' : 'Vice Chairman',
        'Major': 'Computer Science',
        'Team' : 'Machine Learning'
    },

    'Hoyoung': {
        'Role' : 'Chairman',
        'Major': 'Computer Science',
        'Team' : 'Data Visualization'
    }
}

print(students['Hangoo']['Major'])
```
