## if / elif / else statements

If we want to use our code only in a specific case, we can use **if** statement.

```python
score = 60

if score <= 70:
    print("You failed this course!")
```

If there are two, three or more cases to consider, we can use **elif** (else if) and **else** statements.

It is optional to use elif or else statement, but we must use if statement.

```python
score = 90

# Case 1
if score >= 90:
    print("Your grade is A")

# If we can't reach case 1, consider case 2.
elif score >= 80:
    print("Your grade is B")

# If even case 2 is not possible to reach, case 3 is the only option.
else:
    print("You failed this course!")
```

```python
name = 'Jaehong'

if name == 'Hangoo':
    print("Hangoo")

if name == 'Max':
    print("Max")

if name == 'Jaehong':
    print('Jaehong')
```

If statements can also be overlapped. This means that we can have if/else statement inside of another if statement, which can be repeated infinitely.

Important key point is to use **indentation** to set the location of each if/else statement.

```python
score = 85

if score > 50:
    print("At least D")

    if score > 60:
        print("At least C")

        if score > 70:
            print("At least B")

            if score > 80:
                print("At least A")

                if score > 90:
                    print("At least A+")

                else:
                    print("Try to get A+ next time")
```

There can be more than one condition for each if statement. We can use the boolean operation from last week to write less code.

```python
double = 22

if double > 10:
    if double < 30:
        print("Double trouble")

if double > 10 and double < 30:
    print("No trouble")
```

## for / while loops

We use **for** loop to repeat the code for a specific given time.

```python
# for x in y:     x : iterator, y : iterated
        # <code>

# range() function returns a sequence of numbers,
# starting from 0 by default, and increments by 1
for i in range(10):
    print(i)
```

```python
for insta_id in range(3):
    print("dann2lim")
```

```python
# It is possible to specify the starting value for range function
for i in range(1, 5):
    print(i ** 2)
```

```python
# It is possible to specify the increment value for range function
for i in range(1, 5, 2):
    print(i ** 2)
```

```python
# What will this code do?
for i in range(10, 0, -1):
    print(i)
```

We can use **while** loop when we want to repeat the code until a given condition is fullfilled.

```python
# while <condition>:
        # <code>

score = 0

# The code only works when score is less than 10
# When score is greater than or equal to 10, system exits the loop.
while score < 10:
    score += 1
    print(score)

print("We escaped from the loop!")
```

Unfortunately, we can be stuck in while loop forever. This will be mainly logical problems, so keep in mind!

```python
x = 0
while x >= 0:
    x += 1

print("Did we escape?")
```

For loops and while loops are interchangeable.

```python
print("Example of using for loop")
for i in range(5):
    print(i)

print("Example of using while loop")
j = 0
while j < 5:
    print(j)
    j += 1
```

Loops can also be overlapped. We can mix for and while loops too.

```python
something = 10
count1 = 0
count2 = 0

if (something % 2 == 0):
    for i in range(10):
        for j in range(3):
            if something == 8:
                print("Here", count1)
                count1 += 1
            else:
                print("There", count2)
                count2 += 2
```

## Lists

**Lists** are used to store multiple items in a single variable.

```python
# (Name of the list) = [(Elements in the list)]

# We can create an empty list when we don't have any elements in it.
empty_list = []

number_list = [1, 2, 3, 4, 5]

string_list = ["apple", "banana", "tomato", "durian"]

mixed_list = [1, "apple", True, 2.45]
```

To find an element in the list, we can use **index**.

The first item has index [0], and the second item has index [1].

```python
print(number_list[2])
```

When the index is a negative number, the item is chosen from the back of the list.

```python
print(string_list[-1])
print(string_list[len(string_list) - 1])
```

There can be a list inside a list too.

```python
list_in_list = [[1, 2], [3, 4], [5, 6]]

# To access a list inside of a list, we have to set the index at least two times.
print(list_in_list[2][1])
```

```python
'''
 [
 [1, 2]
 [3,4]
 [5,6]
]
'''
```

We can also create a list using for loop.

This is known as **list comprehension**.

```python
list_1  = [1, 2, 3, 4, 5]
print("list_1", list_1)

list_2 = [i for i in range(1, 6)]
print("list_2", list_2)
```


When we want to search an element of a list, we can use for loop.

```python
name_list = ["Jaehong", "Hangoo", "Max"]

for name in name_list:
    if name == "Jaehong":
      print(name)
```

We can use **len()** function to find the length of a list.

```python
print(len(name_list))
```
