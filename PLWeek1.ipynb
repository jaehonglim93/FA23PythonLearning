##Basic Operators

We can use **[ +, -, **, /, //, ***, % ]** to do arithmetics.

````python
num_1 = 4
num_2 = 2

# Change the '+' sign to check the result!
print(num_1 / num_2)


# Modulus -> Gets the remainder
print(14 % 6)

# Floor Division -> Rounds the result down to the nearest integer
print(14 // 6)
````

We can use **[ >, >=, <, <= ]** to do comparison.

````python
# Change the '<' sign to check the result!
print(num_1 < num_2)
````

We can do arithmetics and assignment together by **[ +=, -=, *=, /= ]**.

````python
print(num_1)

# num_1 = num_1 + 3
# Change the '+=' sign to check the result!
num_1 += 5

print(num_1)
````

We can do boolean operations by **[ ==, != ]**.

````python
print(1 + 1 == 2)

print(2 + 2 != 4)
````

We can also do boolean operations by **[ not, or, and ]**.

````python
t = True
f = False

# True when all inputs are true
print(t and f)

# True when at least one input is true
print(t or f)

# True when input is false
print(not t)
````
